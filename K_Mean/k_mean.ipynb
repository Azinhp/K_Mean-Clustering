## creating a dataset

```python
from sklearn.datasets import make_blobs
# Generate a synthetic dataset with 300 samples, 2 features, and 3 clusters
x, y = make_blobs(n_samples=300, n_features=2, centers=3)
```

```python
import matplotlib.pyplot as plt 
# Create a scatter plot of the data points
# x[:, 0] represents the x-coordinates (first feature)
# x[:, 1] represents the y-coordinates (second feature)
plt.scatter(x[:, 0], x[:, 1], color = 'b')
plt.show()
```

## K Mean

```python
from sklearn.cluster import KMeans
```

```python
# Create a KMeans model instance with 3 clusters 
model = KMeans(n_clusters=3)

# Create a KMeans model instance with 3 clusters
model.fit(x)
```

```python
# Get the coordinates of the cluster centers
centers = model.cluster_centers_

# Get the cluster labels assigned to each data point
labels =  model.labels_
```

```python
plt.scatter(x[:, 0], x[:, 1], c = labels)
# Overlay the cluster centers on the scatter plot, marked as red 'x's
plt.scatter(centers[:, 0], centers[:, 1], color = 'red', marker='x')
plt.show()
```
